```python
import pandas as pd
import seaborn as sns

dataframe = pd.read_csv('gun-violence-data_01-2013_03-2018.csv')

dataframe.head()
```

```python
dataframe['date'] = pd.to_datetime(dataframe['date'])
```

```python
dataframe['year'] = pd.DatetimeIndex(dataframe['date']).year

dataframe
```

```python
dataframe['month'] = pd.DatetimeIndex(dataframe['date']).month

dataframe
```

```python
sns.scatterplot(data=dataframe, x='month', y='n_killed', hue='year')
```

```python
sns.scatterplot(data=dataframe, x='year', y='n_killed', hue='month')
```

```python
sns.barplot(data=dataframe, x='year', y='n_killed')
```

```python
sns.barplot(data=dataframe, x='month', y='n_killed')
```

```python
sns.lineplot(data=dataframe, x='month', y='n_killed')
```

```python
sns.lineplot(data=dataframe, x='year', y='n_killed')
```

```python
sns.scatterplot(data=dataframe, x='month', y='n_injured', hue='year')
```

```python
sns.scatterplot(data=dataframe, x='year', y='n_injured', hue='month')
```

```python
sns.lineplot(data=dataframe, x='month', y='n_injured')
```

```python
sns.lineplot(data=dataframe, x='year', y='n_injured')
```
